```python
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
# Importing data
# Laptop
test_dataset = pd.read_excel(r'C:\Users\Iwanw\OneDrive - Cardiff University\Documents\programs\python\python_data\BCA_results.xlsx')
# Uni
#test_dataset = pd.read_excel(r'C:\Users\User\OneDrive - Cardiff University\Documents\programs\python\python_data\BCA_results.xlsx')

# Define the standard curve
standard_concentrations = np.array([0, 2.5, 5, 10, 25, 37.5, 50, 75])
pre_standard_absorbances = test_dataset[[0, 2.5, 5, 10, 25, 37.5, 50, 75]].mean().values

# Define the sample dilutions
sample_dilutions = np.array([0.025, 0.05, 0.1, 0.25, 0.5, 1])
pre_sample_absorbances = test_dataset[[0.025, 0.05, 0.1, 0.25, 0.5, 1]].mean().values

offset = pre_standard_absorbances[0]
standard_absorbances = pre_standard_absorbances - offset
sample_absorbances = pre_sample_absorbances - offset

# Removing 0 ( # if you would like graph to cross 0,0)
#standard_concentrations = standard_concentrations[standard_concentrations != 0]
#standard_absorbances = standard_absorbances[standard_absorbances != 0]

# Fit a polynomial regression model to the standard data
def polynomial_fit(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d
# Regression to calculate concentrations
params, params_covariance = curve_fit(polynomial_fit, standard_absorbances, standard_concentrations)
a, b, c, d = params
# Use the regression equation to calculate the corresponding concentrations for the sample absorbances
sample_concentrations = a * sample_absorbances**3 + b * sample_absorbances**2 + c * sample_absorbances + d
# Correct for the sample dilutions
corrected_sample_concentrations = sample_concentrations / sample_dilutions * 10
print("\nRegression equation: y = {:.10f}x^3 + {:.10f}x^2 + {:.10f}x + {:.10f}".format(a, b, c, d))
print("\n\nSample Concentrations (ug/mL):",corrected_sample_concentrations, "\n")
# Inverted regression to image
params, params_covariance = curve_fit(polynomial_fit, standard_concentrations, standard_absorbances)
a, b, c, d = params

# Plotting
plt.scatter(standard_concentrations, standard_absorbances, color='red')
plt.plot(standard_concentrations, polynomial_fit(standard_concentrations, a, b, c, d), color='orange')
#plt.scatter(sample_concentrations, sample_absorbances, color='blue')
plt.xlabel('Standard Concentration')
plt.ylabel('Standard Absorbance')
plt.xlim([np.min(0), np.max(sample_concentrations * 1.1)])
plt.ylim([np.min(0), np.max(sample_absorbances * 1.1)])
for i, (c, a) in enumerate(zip(sample_concentrations, sample_absorbances)):
    plt.vlines(c, 0, a, color='green', linestyle='--')
    plt.hlines(a, 0, c, color='green', linestyle='--')
current_time = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = "BCA_results_" + current_time + ".pdf"
plt.savefig(filename)
filename1 = "BCA_results_" + current_time + ".jpg"
plt.savefig(filename1)
plt.show()
```
